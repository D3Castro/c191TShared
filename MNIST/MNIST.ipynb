```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt, matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm #support vector machine
from sklearn.decomposition import PCA #principal component analysis 
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(train.shape) #Show (#rows, #columns)
train.head() 
```

```python
#Save labels and drop them from train for ease
labels = train['label']
imgtrain = train.drop("label",axis=1)
imgtrain.head()
```

```python
#Display a random number to visualize data
i = 49
img=imgtrain.iloc[i].as_matrix().reshape((28,28))
plt.imshow(img,cmap='binary')
plt.title(labels.iloc[i])
```

```python
#Fix color values to binary black/white
imgtrain[imgtrain>0] = 1
test[test>0] = 1

img=imgtrain.iloc[i].as_matrix().reshape((28,28))
plt.imshow(img,cmap='binary')
plt.title(labels.iloc[i])
```

```python
train_imgs, test_imgs, train_labels, test_labels = train_test_split(imgtrain, labels, train_size =0.8, random_state=0)
train_imgs[train_imgs>0] = 1

#set up the principal component analysis 
pca = PCA(n_components=16, svd_solver='randomized',whiten=True).fit(train_imgs)
#run PCA on the images
pcatrain = pca.transform(train_imgs)

#set up the support vector machine classifier
classifier = svm.SVC()
classifier.fit(pcatrain, train_labels)
classifier.score(pca.transform(test_imgs),test_labels)
```

```python
#write answers to submission file
testPred = classifier.predict(pca.transform(test))

test['Label'] = pd.Series(testPred)
test['ImageId'] = test.index+1

result = test[['ImageId', 'Label']]
result.to_csv('submission.csv', index=False)
```
